```python
import json

with open('result.json', 'r') as file:
    data = json.load(file)

data = data['data']

nodes = []
for item, value in data.items():
    if isinstance(value, list):
        if item != 'edges':
            nodes += value
    if isinstance(value, dict):
        if item != 'edges':
            nodes.append(value)

print(f"flowchart LR")
for node in nodes:
    name = list(node.keys())[2]
    print(f"    {node['id']}({node['label']} - {node[name]})")
for edge in data['edges']:
    print(f"    {edge['from']} --> {edge['to']}")
            
```

```python
with open('data.json', 'r') as file:
    data = json.load(file)

nodes = []
dict_nodes = data['nodes']
keys = list(dict_nodes.keys())
for key in keys:
    nodes += dict_nodes[key]
edges = data['edges']

print(f"flowchart TD")
for node in nodes:
    dict_attr = list(node.keys())[2]
    print(f"    {node['~id']}({node['~label']} - {node[dict_attr]})")
for edge in edges:
    print(f"    {edge['~from']} --> |{edge['~label']}| {edge['~to']}")
```
